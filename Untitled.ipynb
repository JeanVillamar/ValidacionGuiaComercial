# ABERTURA ARCHIVO EXTENDIDO

```python
workbook = load_workbook(filename= "archivo2.xlsx")
#sheet_names = workbook.sheetnames #obtner las hojas del excel
worksheet = workbook["archivo"] #Acceder a la hoja "archivo"
```

### OBTENCION DE DATOS EXTENDIDO

Documentacion
Indexaciones

[6]  -> fecha de inicio

[7]  -> fecha de fin

[8]  -> promo_description = en las cuales solo nos interesan las descripciones que contienen los siguientes caracteres "AGC","GCA","GC" dado que
                            representan especificamente las guia comerciales gc:guia comercial; agc:anexo guia comercial

[23] -> reward_itemcode = id de la promoción

[24] -> producto

[27] -> reward_type = el tipo 

[28] -> porcentaje

[32] -> farmacias                             

[57] -> proveedor

```python
#Colocamos el min_row en 2 para no contar con el encabezado
diccItem = {}
#input de branches [32]
#input promo description[8]
#promo_description = "GCA BNF 21F+9F SIEGFRIED CCD"
#diccItem["Encabezado"]= ["TIPO PROMOCION","PROVEEDOR","reward_itemcode","PRODUCTO_PARTICIPANTE","PORCENTAJE","INICIO","FIN","BRANCHES"]

for row in worksheet.iter_rows(min_row=2, min_col=1, max_col=58):        
    if(("AGC" in row[8].value or "GCA" in row[8].value or "GC" in row[8].value) and row[27].value == "Porcentaje de Descuento" ):
        if(diccItem.get(row[23].value) is None ):
            diccItem[row[23].value] = [row[27].value, row[57].value,row[23].value,row[24].value,row[28].value,
                                       row[6].value,row[7].value,row[32].value]            
            #for cell in row: #acceder a cada celda
                
        elif(row[23].value is not None):
            #Igual reward_itemcode       
            diccItem[row[23].value] += [row[27].value, row[57].value,row[23].value,row[24].value,row[28].value, 
                                        row[6].value,row[7].value,row[32].value]

```

### ESCRITURA DE ARCHIVO EXCEL EXTENDIDO 

```python

#Creamos el archivo excel vacio
workbook = Workbook()
#Obtenmos la hoja activa
hoja = workbook.active
n = 8

for clave, valor in diccItem.items():
    cantidad = len(valor)
    a = 0
    
    if(cantidad > n):
        for i in range(1 , int(cantidad/n) + 1):
            limite = i*n
            if(len(valor[a: limite+1]) > 0):
                print(valor[a:limite+1])
                hoja.append(valor[a:limite+1])     
            a = limite  
    else:
        hoja.append(valor[a:n])

#guardamos el archivo excel
#archfinal.save(filename = "final.xlsx", overwrite = True)  por alguna extraña razón no reconoce el argumento overwrite
workbook.save(filename = "extendido.xlsx")
```
